Importar librerías

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Importar dataset

```python
df = pd.read_csv('./  Salary_dataset.csv')
```

Calcular las sumas necesarias para los coeficientes de la regresión

```python
n = len(df['YearsExperience'])
X = df['YearsExperience']
Y = df['Salary']
sum_x = X.sum()
sum_y = Y.sum()
sum_x_squared = (X ** 2).sum()
sum_xy = (X * Y).sum()
```

Calcular los coeficientes de la regresión lineal

```python
slope = (n * sum_xy - sum_x * sum_y) / (n * sum_x_squared - sum_x ** 2)
intercept = (sum_y - slope * sum_x) / n
```

Función para hacer las predicciones

```python
def predict(x):
    return slope * x + intercept
```

Datos aleatorios para hacer las pruebas

```python
x_test = np.random.rand(len(X), 1) * 11
y_test = np.zeros((len(X),1))

for index, value in enumerate(x_test):
  y_test[index] = predict(value)
```

Gráficar

```python
plt.scatter(X, Y, color='b')
plt.plot(x_test, y_test, color='r')
plt.xlabel('Variable independiente')
plt.ylabel('Variable dependiente')
plt.show()
```
